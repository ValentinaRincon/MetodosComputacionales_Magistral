```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from tqdm import tqdm
```

```python
# Random walk
N = 3000

x = np.zeros(N)
y = np.zeros(N)

for i in range(1,N):
    theta = 2.*np.pi*np.random.rand()
    x[i] = x[i-1] + np.cos(theta)
    y[i] = y[i-1] + np.sin(theta)
```

```python
# Definamos funcion de animacion
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

def init():
    ax.set_xlabel('x[m]')
    ax.set_ylabel('y[m]')
    ax.set_xlim(-10,10)
    ax.set_ylim(-10,10)
    
# Definamos funcion de actualizacion
def Update(i):
    #ax.clear()
    init()
    ax.plot(x[:i],y[:i],'k-o',markersize=5)
    ax1.plot(x[:i],color='b')
    
    #plt.close()

Animation = anim.FuncAnimation(fig, Update, frames=N, init_func=init, interval=100) 
```
